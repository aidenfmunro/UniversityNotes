```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
torr_40 = pd.read_csv('40tor.csv')
torr_40_r = pd.read_csv('40tor_r.csv')
torr_60 = pd.read_csv('60tor.csv')
torr_80 = pd.read_csv('80tor.csv')
```

```python
def convert(dataframe):
    dataframe = dataframe.to_numpy()
    dataframe = dataframe.reshape(-1)
    return dataframe[::2], dataframe[1::2]
```

```python
torr_40_time, torr_40_volt = convert(torr_40)
torr_40_r_time, torr_40_r_volt = convert(torr_40_r)
torr_60_time, torr_60_volt = convert(torr_60)
torr_80_time, torr_80_volt = convert(torr_80)
```
